### Creating new features

#### In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. Use .apply() with a custom lambda expression to create a new column called "Reason" that contains this string value. 



#### For example, if the title column value is EMS: BACK PAINS/INJURY, the Reason column value would be EMS.

```python
df['Reason'] = df['title'].apply(lambda s:s.split(':')[0])
df['Reason'].head()
```

#### What is the most common Reason for a 911 call based off of this new column?

```python
df['Reason'].value_counts()
```

#### Now use seaborn to create a countplot of 911 calls by Reason.

```python
sns.countplot(x='Reason', data = df)
```

#### What is the data type of the objects in the timeStamp column?

```python
type(df['timeStamp'].iloc[0])
```

#### Convert the column from strings to DateTime objects.

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
type(df['timeStamp'])
```

#### Grab specific attributes from a Datetime object by calling them. For example: 

```python
time = df['timeStamp'].iloc[0]
time.hour
```

#### Use .apply() to create 3 new columns called Hour, Month, and Day of Week based off of the TimeStamp column.

```python
df['Hour'] = df['timeStamp'].apply(lambda time:time.hour)
df['Month'] = df['timeStamp'].apply(lambda time:time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
df.sample()
```

#### The Day of Week is an integer 0-6. Use the .map() with this dictionary to map the actual string names to the day of the week:

```python
dmap = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].apply(lambda int:dmap[int])
```

#### Use seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column.

```python
sns.countplot(x='Day of Week', hue='Reason', data=df)
plt.legend(bbox_to_anchor=(1,1))
```

#### Now do the same for Month

```python
sns.countplot(x='Month', hue='Reason', data=df)
plt.legend(bbox_to_anchor=(1, 1))
```

#### Creat a new column called 'Date' that contains the date from the timeStamp column. You will need to use apply along with the .date() method.

```python
df['Date'] = df['timeStamp'].apply(lambda time:time.date())
```

```python
df.head()
```

#### Now groupby this Date Column with the count() aggregate and create a plot of counts of 911 calls.

```python
df.groupby(by='Date').count()['lat'].plot()
plt.tight_layout()
```

#### Recreat this plot but create 3 separate plots with each plot representing a Reason for 911 call.

```python
df[df['Reason']=='Traffic'].groupby(by='Date').count()['lat'].plot()
plt.title('Traffic')
```

```python
df[df['Reason']=='Fire'].groupby(by='Date').count()['lat'].plot()
plt.title('Fire')
```

```python
df[df['Reason']=='EMS'].groupby(by='Date').count()['lat'].plot()
plt.title('EMS')
```

#### Restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. Combine groupby with an unstack method.

```python
dfGrid = df.groupby(by=['Day of Week', 'Hour']).count()['lat'].unstack()
dfGrid = dfGrid.loc[['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']]
dfGrid
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dfGrid, cmap='viridis')
```

#### Creae a clustermap using this DataFrame

```python
sns.clustermap(dfGrid, cmap='viridis')
```
